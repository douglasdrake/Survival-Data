## Summary

Four treatments (Capomulin, Infubinol, Ketapril, and Placebo) were considered over the course of 45 days with 25 mice per treatment.  Evaluations of tumor volume, number of metastatic sites, and number of surviving mice were made every five days. Plots exploring these observations are given below.  From these plots, we observe the following:

* __Tumor Volume__: For all four treatments, average tumor volume is approximately linearly related to days of treatment.  Capomulin is the only treatment that results in the average tumor volume decreasing over time.  For Capomulin, each 10 days of treatment results in the average volume decreasing by approximately 2 mm$^3$. The other three treatments resulted in the average tumor volume increasing over time.  For each 10 days of treatment, the average tumor volume under Infubinol, Ketapril, and placebo increased by 5 mm$^3$, 6 mm$^3$, and 5 mm$^3$, respectively.  Infubinol and Ketapril are nearly indistinguishable from the placebo with respect to tumor volume.  
* __Metastatic Sites__:  The average number of metastatic sites increased (approximately linearly) under all four treatments.  The treatments ranked from lowest to highest with respect to rate of increase in metastatic sites are: Capomulin, Infubinol, Ketapril and Placebo.  After 45 days of treatment, the average number of metastatic sites for Capomulin, Infubinol, Ketapril, and Placebo were 1.5, 2.1, 3.4, and 3.3, respectively.
* __Survival Percent__:  As expected, the percent of surviving mice decreases over the treatment duration for all four treatments. The survival percentage under Capomulin is highest among the four treatments and decreases at the slowest rate across the 45 days of treatment.  At day 45, the percentages of surviving mice for Capomulin, Ketapril, Placebo, and Infubinol are 84%, 44%, 43%, and 36%, respectively.

While we have not performed formal tests of significance, Capomulin appears to offer the best outcomes based on the three metrics (tumor volume, metastatic sites, and survival percentage) and appears to differ in a significant manner from the other three treatments.  With respect to tumor volume and survival percentage, Infubinol and Ketapril are nearly indistinguishable from the placebo.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Set paths to CSV files and give a function to load the data:
MOUSE_DIR = "data"
MOUSE_FILE = "mouse_drug_data.csv"
MOUSE_CSV = os.path.join("", MOUSE_DIR, MOUSE_FILE)
TRIAL_FILE = "clinicaltrial_data.csv"
TRIAL_CSV = os.path.join("", MOUSE_DIR, TRIAL_FILE)

# The following function will read the Mouse and Drug Data and 
# the Clinical Trial Data and then combine the data into a single dataset
def load_mouse_data(mouse_file = MOUSE_CSV, trial_file = TRIAL_CSV):
    mouse_df = pd.read_csv(mouse_file)
    trial_df = pd.read_csv(trial_file)

    # the following merge would match the output on GitLab
    # which sorts the output based on timepoints
    # combined_df = pd.merge(trial_df, mouse_df, on = "Mouse ID", how = 'left')
    # the following gives a similar table arranged on Mouse ID
    combined_df = pd.merge(trial_df, mouse_df, on = "Mouse ID")
    
    return combined_df

# Display the data table for preview
mouse_df = load_mouse_data()

mouse_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
grouped_df = mouse_df.groupby(['Drug', 'Timepoint'], as_index=False)
grouped_volume = grouped_df['Tumor Volume (mm3)'].mean()

# Preview DataFrame
grouped_volume
```

```python
# Import dependency for standard error calculations
from scipy.stats import sem
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
grouped_sems = grouped_df['Tumor Volume (mm3)'].agg(['sem'])
grouped_sems = grouped_sems.reset_index()
grouped_sems = grouped_sems.rename(columns = {'sem' : 'Tumor Volume (mm3)'})

# Preview the DataFrame
grouped_sems.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
means_df = grouped_volume.pivot(index = 'Timepoint', columns='Drug', values='Tumor Volume (mm3)')
sems_df = grouped_sems.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview
means_df.head()
#sems_df.head()
```

```python
# Generate the Plot (with Error Bars)

# We will write a function that takes the two data frames and a list of 
# drug names to plot

def plot_tumor_data (m_df, se_df, drugs, colors, symbols,\
                     xlabel, ylabel, title, filename):
    """ Plot an errorbar plot for a subset of the drugs in the experiment
    
    Args:
        m_df: a DataFrame of mean values versus time for each drug
        se_df: a DataFrame of the standard errors of the means for each drug or `None` for no errorbars
        drugs: a list of drug names to be used in the plot.
        colors: a dict of the form {"drugname" : "color"} for plotting
        symbols: a dict of the form {"drugname" : "symbol"} for plotting
        xlabel: a string for the xlabel
        ylabel: a string for the ylabel
        title: a string for the title
        filename: a file for saving the figure
    
    Returns:
        the return values of the plt.subplots() function
    """
    fig, ax = plt.subplots()

    for drug in drugs:
        means = m_df[drug]
        if se_df is not None:
            sems = se_df[drug]
        else:
            sems = 0
        x_axis = m_df.index
        ax.errorbar(x_axis, means, yerr=sems,\
                    linestyle="--", linewidth=0.5,\
                    fmt=symbols[drug], color=colors[drug],\
                    alpha=0.7, label=drug)
        
    #ax.set_xlim(-0.5, len(means))
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.grid()
    plt.legend(loc="best", fontsize="small", fancybox=True)

    fig.savefig(filename)
    plt.show()

    return fig, ax

# suppress the return values
_ = plot_tumor_data(means_df, sems_df,\
                    ["Capomulin", "Infubinol", "Ketapril", "Placebo"],\
                    {"Capomulin" : "red", "Infubinol" : "blue", "Ketapril" : "green", "Placebo" : "black"},\
                    {"Capomulin" : "o", "Infubinol" : "^", "Ketapril" : "s", "Placebo" : "d"},\
                    "Time (Days)", "Tumor Volume (mm3)", "Tumor Response to Treatment",\
                    "tumorvolume.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
grouped_df = mouse_df.groupby(['Drug', 'Timepoint'])
grouped_metastatic = grouped_df['Metastatic Sites'].agg('mean')

# Convert to DataFrame
grouped_metastatic = pd.DataFrame(grouped_metastatic)

# Preview DataFrame
grouped_metastatic.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint
# Convert to DataFrame - this is done by the pd.DataFrame.agg() method
grouped_ms_sems = grouped_df['Metastatic Sites'].agg(['sem'])
grouped_ms_sems = grouped_ms_sems.rename(columns = {'sem' : 'Metastatic Sites'})

# Preview the DataFrame
grouped_ms_sems.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Note that we need to eliminate the hierarchical indices before pivoting
grouped_metastatic = grouped_metastatic.reset_index()
grouped_ms_sems = grouped_ms_sems.reset_index()

ms_means_df = grouped_metastatic.pivot(index = 'Timepoint', columns='Drug', values='Metastatic Sites')
ms_sems_df = grouped_ms_sems.pivot(index = 'Timepoint', columns = 'Drug', values = 'Metastatic Sites')

# Preview
ms_means_df.head()
#ms_sems_df.head()
# Note that the starter notebook displays the tumor volume data
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure

# Suppress the return values
_ = plot_tumor_data(ms_means_df, ms_sems_df,\
                    ["Capomulin", "Infubinol", "Ketapril", "Placebo"],\
                    {"Capomulin" : "red", "Infubinol" : "blue", "Ketapril" : "green", "Placebo" : "black"},\
                    {"Capomulin" : "o", "Infubinol" : "^", "Ketapril" : "s", "Placebo" : "d"},\
                    "Treatment Duration (Days)", "Met. Sites", "Metastatic Spread During Treatment",\
                    "metatstatic.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
grouped_df = mouse_df.groupby(['Drug', 'Timepoint'], as_index=False)
grouped_survival = grouped_df['Mouse ID'].count()
grouped_survival = grouped_survival.rename(columns = {"Mouse ID" : "Mouse Count"})

# Preview DataFrame
grouped_survival.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_df = grouped_survival.pivot(index = 'Timepoint', columns = 'Drug', values = 'Mouse Count')

# Preview the Data Frame
survival_df.head()
```

```python
# We will use the following helper function to compute the survival percentage 
# before passing it to plotting function
def compute_survival_percent (df):
    """Returns the dataframe of counts with survival percents."""
    # A helper function
    def percent_survival (x):
        return pd.Series(100 * x / x[0])

    percent_survival_df = df.copy()
    for column in df.columns:
        percent_survival_df[column] = percent_survival(df[column])

    return percent_survival_df

percent_survival_df = compute_survival_percent(survival_df)
#percent_survival_df
```

```python
# Generate the Plot (Accounting for percentages)
# Save the Figure
# Show the Figure

# We use the plotting function written above:
# suppress the return values
_ = plot_tumor_data(percent_survival_df, None,\
                    ["Capomulin", "Infubinol", "Ketapril", "Placebo"],\
                    {"Capomulin" : "red", "Infubinol" : "blue", "Ketapril" : "green", "Placebo" : "black"},\
                    {"Capomulin" : "o", "Infubinol" : "^", "Ketapril" : "s", "Placebo" : "d"},\
                    "Time (Days)", "Survival Rate (%)", "Survival During Treatment",\
                    "survival.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

def percent_change (vol_df):
    
    # The series in the tumor volume dataframe are
    # indexed by the time points so we can't use '-1'
    # as an index value to get the last element.
    # A helper function - this will work with any Series
    f = lambda x: 100 * (x.iloc[-1] - x.iloc[0]) / x.iloc[0]
    
    return vol_df.apply(f)
 
percent_change_series = percent_change(means_df)    

# Display the data to confirm
percent_change_series
```

```python
# Use the pandas.plot method and then adjust the plot object with setter methods
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
ax = percent_change_series[drugs].plot(kind="bar", rot=0, grid=True, align="edge",\
                                      color=["green", "red", "red", "red"], width=1, alpha=0.7)
ax.set_xlabel("")
ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change Over 45 Day Treatment")
ax.set_xticks(np.arange(1,5))
ax.set_yticks(np.arange(-20, 80, 20))
ax.set_xlim(-0.2, 4.2)
ax.set_ylim(-30, 70)

# Now add the percentages to the plot
percents =  percent_change_series[drugs]
x_axis = np.arange(0.5, 4.5, 1)
for x, percent in zip(x_axis, percents):
    if percent < 0:
        y = -5
    else:
        y = 5
    ax.text(x, y, f"{percent:.0f}%", horizontalalignment="center", color="white")

plt.savefig("tumorchange.png")
```

### Additional code used to compare the four treatments.
This information is used in the summary.

```python
from scipy.stats import linregress

def compute_slopes(df, drugs):
    """Compute the slopes for the change over time for each drug."""
    
    slopes = []
    x = df.index
    
    for drug in drugs:
        y = df[drug]
        (slope,_,_,_,_) = linregress(x, y)
        slopes.append(slope)
    
    return slopes
```

```python
# Look at the slopes for tumor volume
import math
drugs = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
means_slopes = compute_slopes(means_df, drugs)

for drug, slope in zip(drugs, means_slopes):
    if slope > 0:
        change = "increase"
    else:
        change = "decrease"
    strout = f"{drug}: An increase of 10 treatment days results in an average {change} of {math.fabs(10*slope):2.1f} mm3 in tumor volume."
    print(strout)
```

```python
# Look at the slopes for metastatic spread
import math
drugs = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']
ms_means_slopes = compute_slopes(ms_means_df, drugs)

for drug, slope in zip(drugs, ms_means_slopes):
    if slope > 0:
        change = "increase"
    else:
        change = "decrease"
    strout = f"{drug}: An increase of 10 treatment days results in an average {change} of {math.fabs(10*slope):2.1f} metastatic sites."
    print(strout)
```
